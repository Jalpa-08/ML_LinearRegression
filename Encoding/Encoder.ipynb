```python
# include only object type datatype only
my_object_df = df.select_dtypes(include='object') #return all string/ categorical columns
my_object_df
```

```python
# Consider only neumerical values
my_numeric_df = df.select_dtypes(exclude='object') #return all numeric columns
my_numeric_df
```

```python
my_numeric_df = my_numeric_df.drop('Unnamed: 0', axis=1)
my_numeric_df.head()
```

```python
df_object_dummies = pd.get_dummies(my_object_df, drop_first = True) #encoded dataframe
df_object_dummies
```

```python
final_df = pd.concat([my_numeric_df, df_object_dummies], axis=1)
final_df
```

```python
X =final_df.drop('SalePrice', axis=1)
X.head()
```

```python
y =final_df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_model = ElasticNet(max_iter = 20000000)
```

```python
param_grid = {'alpha': [0.1,1,5,10,100], 'l1_ratio':[0.1, 0.5, 0.7, 0.9, 1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator = base_elastic_model, param_grid = param_grid, scoring='neg_mean_squared_error', cv=5, verbose=2)
```

```python
grid_model.fit(scaled_X_train, y_train)
```
