```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_csv(r'C:\Users\Jalpa\Desktop\Modular-1\Machine-Learning-Models\DATA\Mall_Customers.csv')
df1
```

```python
df =df1.drop(columns = ['Unnamed: 0'] , axis=1)
df
```

```python
from sklearn.cluster import KMeans
```

```python
x = df.iloc[: , 3:]
x
```

```python
wcss = []
for i in range(1,15):
    kmean = KMeans(n_clusters=i , init='k-means++' , random_state=30)
    kmean.fit(x)
    wcss.append(kmean.inertia_)
    
```

```python
wcss
```

```python
plt.plot(range(1,15)  , wcss)
```

```python
kmean1= KMeans(n_clusters=5 ,init = 'k-means++' , random_state=30)
kmean1.fit_predict(x)
```

```python
x['cluster number'] = kmean1.fit_predict(x)
```

```python
x[x['cluster number'] ==3]
```

```python
x[x['cluster number'] ==1]
```

```python
55 , 31
```

```python
kmean1.predict([[69,91]])
```

# Minibatch Kmean

```python
from sklearn.cluster import MiniBatchKMeans
minibatch_kmean = MiniBatchKMeans(n_clusters=5)
minibatch_kmean.fit(x)
```

```python
minibatch_kmean.predict([[55,31 ,4]])
```

# DBSCAN

```python
from  sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=1, min_samples=3)
dbscan.fit(x)
```

```python
x
```

```python
dbscan.labels_
```

```python
len(set(dbscan.labels_))
```

```python
x['cluster number '] = dbscan.labels_
```

```python
x
```

```python
from sklearn import metrics
```

```python
true_label = x['cluster number']
predicted_label = x['cluster number']
metrics.adjusted_rand_score(true_label,predicted_label)
```

```python
metrics.adjusted_rand_score(x['cluster number'],dbscan.labels_)
```

```python
metrics.adjusted_rand_score(true_label,predicted_label)
```

```python
metrics.jaccard_score(true_label,predicted_label, average='macro')
```
